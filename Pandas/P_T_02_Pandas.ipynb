```python
# numpy: numerical computing library for arrays and math operations
import pandas as pd

# numpy: numerical computing library for arrays and math operations
import numpy as np

# requests: library to make HTTP requests and fetch data from URLs
import requests

```

```python
# StringIO: treats a string as a file object (for text data like CSV)
# BytesIO: treats bytes as a file object (for binary data like Excel)
from io import StringIO, BytesIO
```

```python
data = ('col1,col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
        'c,d,3')
```

```python
# type(data): returns the data type/class of an object
type(data)
```

```python
# StringIO(): creates an in-memory file-like object from a string
# used to convert string data into readable format for pandas
StringIO()
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from specific columns
df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in['COL1','COL3'])
```

```python
df
```

```python
# df.to_csv(): exports DataFrame to a CSV file
df.to_csv('Test(pt6).csv')
```

```python
data = ('a,b,c\n'
        '4,apple,bat,\n'
        '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data),index_col=False)
```

```python
## Combining usecols and index_col
data = ('a,b,c\n'
        '4,apple,bat,\n'
        '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data), usecols=['b', 'c'],index_col=False)
```

```python
## Quoting and Escape Characters. Very useful in NlP
data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data), escapechar='\\')
```

```python
## URL to CSV

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

url = ('https://download.bls.gov/pub/time.series/cu/cu.item')
response = requests.get(url, headers=headers)
df = pd.read_csv(StringIO(response.text), sep='\t')
print(df.head())
```
