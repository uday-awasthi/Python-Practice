### Read Json to CSV

JSON (JavaScript Object Notation) is a lightweight data format used for storing and exchanging data.
Using Pandas, we can read a JSON file and convert it into a CSV file for easy analysis and storage.

```python
data = '''
{"employee_name":"James","email":"james@gmail.com","job_profile":[{"title1":"Team Lead","title2":"Sr. Developer"}]}
{"employee_name":"James","email":"james@gmail.com","job_profile":[{"title1":"Team Lead","title2":"Sr. Developer"}]}
'''

df1 = pd.read_json(StringIO(data), lines=True)
print(df1)
```

### Reading HTML content

Pandas can read tables directly from HTML web pages using the read_html() function.
It extracts all tables available on a webpage and stores them as DataFrames.

```python
url = "https://www.fdic.gov/bank/individual/failed/banklist.html"
dfs = pd.read_html(url)
```

```python
dfs[0]
```

```python
url = "https://en.wikipedia.org/wiki/Mobile_country_code"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

dfs = pd.read_html(StringIO(response.text), match="Country", header=1)
```

```python
dfs[0]
```
