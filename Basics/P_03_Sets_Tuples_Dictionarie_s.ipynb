### SETS
A set is and unordered collection data type that is iterable, mutable, and has no duplicate elemenst. Pythons set class represents the mathematical notion of a set. This is based on a data structure known as a hash table

```python
## Defining an empty set
set_var = set()
print(set_var)
print(type(set_var))
```

```python
set_var={1,2,3,4,3}
```

```python
set_var
```

```python
set_var = {"Avengers","Ironman","Hitman"}
print(set_var)
type(set_var)
```

```python
##Inbuilt function in sets
set_var.add("Hulk")
```

```python
print(set_var)
```

```python
set1= {"Avengers","Ironman",'Hitman'}
set2= {"Avengers","Ironman",'Hitman','Hulk2'}
```

```python
set2.intersection_update(set1)
```

```python
##Difference
set2.difference(set1)
```

```python
set2
```

## Dictionaries
A dictionary is a collection which is unordered, changeable, and indexed. In Python dictionaries are written with curly brackets, and they have keys and values.

```python
dic={}
```

```python
type(dic)
```

```python
set_ex={1,2,3,4,5}
```

```python
type(set_ex)
```

```python
## let create a dictionary
my_dict={"Car1": "Audi", "Car2":"BMW","Car3":"Mercedes Benz"}
```

```python
type(my_dict)
```

```python
##Access the item values based on keys
my_dict['Car1']
```

```python
#We can even loop throught the dictionaries keys
for x in my_dict:
    print(x)
```

```python
#We can even loop throught the dictionaries values
for x in my_dict.values():
    print(x)
```

```python
#we can also cheack both keys and values
for x in my_dict.items():
    print(x)
```

```python
## Adding items in dictionaries
my_dict['car4']='Audi 2.0'
```

```python
my_dict
```

```python
my_dict['Car1']='Maruti'
```

```python
my_dict
```

### Nested Dictionary

```python
car1_model={'Mercedes':1960}
car2_model={'Audi':1970}
car3_model={'Ambassador':1980}

car_type={'car1':car1_model,'car2':car2_model,'car3':car3_model}
```

```python
print(car_type)
```

```python
#Accesing the items in the dictionary
print(car_type['car1'])
```

```python
print(car_type['car1']['Mercedes'])
```
