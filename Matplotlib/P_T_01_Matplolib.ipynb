## MatplotLib Tutorial
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embadding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:
. Generally easy to get started for simple plots
. Support for custom labels and texts
. Great control of every element in figure 
. High-quality outputs in many formats
. Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
## Simple Examples

x = np.arange(0,10)
y = np.arange(11,21)
```

## plt scatter

Definition: Creates a scatter plot to show the relationship between two numerical variables using individual data points.

```python
## plotting using matplotlib

## plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
## plt.savefig('Test.png') # It save the graph in this format in same folder

# plt.show() if i have not imported  %matplotlib inline so i have to write plt.show()
```

```python
y = x*x
```

## pltt plot

Definition: Creates a line plot used to visualize trends, patterns, or changes in data over a continuous interval.

```python
## plt plot

plt.plot(x,y,'r*',linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots

## Definition: 
# Subplots allow multiple plots to be displayed within a single figure for easy comparison and analysis.
plt.subplot(2,2,1)
plt.plot(x,y,'r*--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0 , 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sine wave form")

# plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2,1,1)

# Make the first plot
plt.plot(x, y_sin,'r--')
plt.title('Sine')

# set the second subplot as active, and make the second plot .
plt.subplot(2, 1, 2)
plt.plot(x, y_cos,'g--')
plt.title('Cosine')

# Show the figure
plt.show()
```

## Bar plot

A bar plot represents categorical data using rectangular bars whose heights or lengths correspond to the values.

```python
## Bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x, y, align = 'center')
plt.bar(x2, y2, color = 'g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('x axis')
plt.show()
```

## Histograms

A histogram displays the frequency distribution of numerical data by dividing it into intervals called bins.

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=30) ## You can change bins size or no bindefault
plt.title("histogram")
plt.show()
```

## Box Plot using Matplotlib

A box plot summarizes data distribution using the minimum, first quartile, median, third quartile, and maximum values, helping identify outliers.

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False);
```

## Pie Chart

A pie chart is a circular chart divided into slices, where each slice represents a category's proportion of the whole dataset.

```python
##  Data to plot
labels = 'Python', 'C++','Ruby', 'Java'
sizes = [213, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'blue']
explode = (0.4, 0, 0, 0) # explode !st slice

# Plot
plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True
)
plt.axis('equal')
plt.show()
```
